```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
wine = pd.read_csv("wine.csv")
wine
```

```python
wine_data = wine.iloc[:,1:]
wine_data.head()
```

```python
wine.shape
```

```python
wine.info()
```

```python
wine.describe()
```

```python
wine.isna().sum()
```

```python
wines = wine.values
wines
```

```python
wine_norm = scale(wines)
wine_norm
```

**Heirarchical Clustering**

```python
# create dendrogram
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#Create clusters
hc = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
hc
```

```python
# Save Clusters for chart
Y_hc = hc.fit_predict(wine_norm)
Y_hc
```

```python
Clusters = pd.DataFrame(Y_hc,columns=['Clusters'])
Clusters
```

```python
Clusters.value_counts()
```

```python
wine['h_clusterid'] = Y_hc
wine
```

**K_Means**

```python
#Wcss (Within Clusters sum of squares)
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
    
print(wcss)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Builder Clusters Algorithm
clusters_new = KMeans(3,random_state=0)
clusters_new.fit(wine_norm)
```

```python
clusters_new.labels_
```

```python
# Assign clusters to the data set
wine['clusterid_km'] = clusters_new.labels_
wine
```

```python
#These are standardized values
clusters_new.cluster_centers_
```

```python
wine
```

```python
wine['clusterid_km'].value_counts()
```

**PCA**

```python
pca = PCA()
pca_wine = pca.fit_transform(wine_norm)
pca_wine
```

```python
pca = PCA(n_components=12)
pca_value = pca.fit_transform(wine_norm)
pca_value
```

```python
# The amount Of variance that each PCA explains
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative Variance
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var,color='blue')
```

```python
plt.plot(var1,color='green')
```
