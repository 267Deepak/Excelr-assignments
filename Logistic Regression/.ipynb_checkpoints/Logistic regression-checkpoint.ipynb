```python
# One hot Encoding Of categorical variables
df1 = pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
df1
```

```python
pd.set_option("display.max.columns",None)
df1
```

```python
df1.info()
```

```python
#customs Binary encoding of binary o/p variables
df1['default'] = np.where(df1["default"].str.contains("yes"),1,0)
df1['housing'] = np.where(df1["housing"].str.contains("yes"),1,0)
df1["loan"] = np.where(df1["loan"].str.contains("yes"),1,0)
df1["y"] = np.where(df1["y"].str.contains("yes"),1,0)
df1
```

```python
# Find & rwplace encoding for month categorical variables
df1["month"].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
df1=df1.replace(order)
df1
```

## Model Building

```python
# Dividing our data into output variables
x=pd.concat([df1.iloc[:,0:11],df1.iloc[:,12:]],axis=1)
y=df1.iloc[:,11]
```

```python
# Logistic Regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
# predict for x dataset
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39156+1234)/(39156+766+4055+1234)
```

```python
# as accuracy = 0.8933 which is greater than 0.5:,thus[:,1],Threshould value<0.5=0 
classifier.predict_proba(x)[:,1]
```

```python
# ROC curve plotting and finding AUC value
fpr,tpr,thresholds = roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc = roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive rate or [1 - True Negative rate]')
plt.ylabel('True  Positive Rate')
plt.show()
print('auc accuracy:',auc)
```
