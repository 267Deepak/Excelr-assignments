```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy.polynomial.polynomial import polyfit
from sklearn.linear_model import LinearRegression
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("D:\Excelr assignments\Simple  Linear Regression\data\delivery_time.csv")
df
```

```python
df.info()
```

```python
df.corr()
```

```python
x = df['Sorting Time']
y = df['Delivery Time']
```

```python
sns.pairplot(df, kind= 'reg')
plt.show()
```

```python
model = sm.OLS(y, x).fit()
prediction = model.predict(x)
prediction
```

```python
model.summary()
```

```python
x_log = np.log(df['Sorting Time'])
print(x_log)
```

```python
# now I will build new model with log transformed x data
model1 = sm.OLS(y, x_log).fit()
pred1 = model1.predict(x_log)
print(pred1)
```

```python
model1.summary()
```

```python
y_log = np.log(df['Delivery Time'])
y_log
```

```python
model2 = sm.OLS(y_log, x).fit()
pred2 = model2.predict(x)
model2.summary()
```

```python
model3 = sm.OLS(y_log, x_log).fit()
pred3 = model3.predict(x_log)
print(pred3)
```

```python
model3.summary()
```

```python
x_sqrt = np.sqrt(df['Sorting Time'])
model4 = sm.OLS(y, x_sqrt).fit()
pred4 = model4.predict(x_sqrt)
print(pred4)
model4.summary()
```

```python
y_sqrt = np.sqrt(df['Delivery Time'])
model5 = sm.OLS(y_sqrt,x).fit()
pred5 = model5.predict(x_sqrt)
print(pred5)
model5.summary()
```

```python
model6 = sm.OLS(y_sqrt, x_sqrt).fit()
pred6 = model6.predict(x_sqrt)
model6.summary()
```
