```python
# Model  Building
model1 = smf.ols("PROFIT~RDS+ADS+MS",data=df1).fit()
```

```python
model1.params
```

```python
model1.tvalues , np.round(model1.pvalues,5)
```

```python
model1.rsquared , model1.rsquared_adj
```

```python
# Build SLR and MLR models for insignificant variables 'ADS' and 'MS'
# Also find their tvalues and pvalues

SLR = smf.ols("PROFIT~MS",data=df1).fit()
SLR.tvalues,SLR.pvalues
```

```python
SLR1=smf.ols("PROFIT~ADS",data=df1).fit()
SLR1.tvalues,SLR1.pvalues
```

```python
MLR = smf.ols("PROFIT~ADS+MS",data=df1).fit()
MLR.tvalues,MLR.pvalues
```

```python
# Model Validation
# Collinearity Check.
```

```python
rsq1 = smf.ols("RDS~ADS+MS",data=df1).fit().rsquared
vif1 = 1/(1-rsq1)

rsq2 = smf.ols("ADS~RDS+MS",data=df1).fit().rsquared
vif2 = 1/(1-rsq2)

rsq3 = smf.ols("MS~RDS+ADS",data=df1).fit().rsquared
vif3 = 1/(1-rsq3)

data = {"Variables":["RDS","ADS","MS"],"Vif":[vif1,vif2,vif3]}
Vif_data = pd.DataFrame(data)
Vif_data
```

```python
# There is No collinearty,All Variables are in regression equations
```

```python
# Residual Analysis
sm.qqplot(model1.resid,line ='q')
plt.title("Normal qq plot of residuals")
plt.show()
```

```python
list(np.where(model1.resid<-30000))
```

```python
# Test  for Homoscedasicity or Hetrosceddasticity
def standard_values(vals):return(vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model1.fittedvalues),standard_values(model1.resid))
plt.title("Residual Plot")
plt.xlabel("Standardized fitted values")
plt.ylabel("Standardized residual values")
plt.show()
```

```python
# Test Of errors or residuals vs independent x variables or predictions
# Using Residual Regression
```

```python
fig2 = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model1,"ADS",fig=fig2)
plt.show()
```

```python
# Model Deletion Diagnostics (checking Outliers or Influencers)
# Cook Distance

(c,_)= model1.get_influence().cooks_distance
c
```

```python
fig4 = plt.figure(figsize=(20,7))
plt.stem(np.arange(len(df1)),np.round(c,5))
plt.xlabel("Row Inex")
plt.ylabel("cooks Distance")
plt.show()
```

```python
np.argmax(c),np.max(c)
```

```python
influence_plot(model1)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n
# k = no.of features/columns & n = no. of datapoints
k=df1.shape[1]
n=df1.shape[0]
leaverage_cutoff = (3*(k+1))/n
leaverage_cutoff
```

```python
df1[df1.index.isin([49])]
```

```python
# improve the model
df2 = df1.drop(df1.index[[49]],axis=0).reset_index(drop=True)
df2
```

```python
# model deletion Diagnostics and Final Model
```

```python
while np.max(c)>0.5 :
    model=smf.ols("PROFIT~RDS+ADS+MS",data=df2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    data2=df2.drop(df2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    data2
else:
    final_model=smf.ols("PROFIT~RDS+ADS+MS",data=df2).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
