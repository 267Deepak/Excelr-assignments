## Finding Correlation between the Variables in the data

```python
cor = zoo1.corr(method='pearson')
cor.style.background_gradient(cmap='coolwarm')
```

```python
# Finding Optional Number of K
X = zoo1.iloc[:,1:17]
y = zoo1.iloc[:,17]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1, stratify=y)
```

```python
k_values = np.arange(1,25)
train_accu=[]
test_accu = []
```

```python
for i,k in enumerate(k_values):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    train_accu.append(knn.score(X_train,y_train))
    test_accu.append(knn.score(X_test,y_test))
    
```

```python
plt.figure(figsize=[13,8])
plt.plot(k_values,test_accu,label="Testing Accuracy")
plt.plot(k_values,train_accu,label="Training Accuracy")
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
y_pred_KNeighborsClassifier = knn.predict(X_test)
```

```python
 # Initialize the "scores" variable as a list
scores = []
cv_scores = []
```

```python
## code to split the data into training and testing sets, and fit the KNeighborsClassifier model
```

```python
for i in range(10):
    y_pred_KNeighborsClassifier = knn.predict(X_test)
    score = accuracy_score(y_pred_KNeighborsClassifier, y_test)
    
    scores.append(score) # Append the accuracy score to the "scores" list

mean_score = np.mean(scores)
print("Mean accuracy score:", mean_score)
```

```python
score_knn=cross_val_score(knn,X,y,cv=10)
```
