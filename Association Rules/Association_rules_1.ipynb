**Prepare rules for the all the data sets 1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values 2) Change the minimum length in apriori algorithm 3) Visulize the obtained rules using different plots**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
book = pd.read_csv('book.csv')
book
```

```python
book.head()
```

# Apriori Algorithm

# 1.Association rules with 10% support and 70% confidence

```python
frequent_itemset = apriori(book,min_support=0.1,use_colnames=True)
frequent_itemset
```

```python
#with 70% Cofidence
rules = association_rules(frequent_itemset,metric="lift",min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rules in selecting the association Transcation
rules[rules.lift>1]
```

```python
# Visulaization of obtained rules 
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

# 2. Association rules with 20% support and 60% confidence

```python
# With 20% support 
frequent_itemset2 = apriori(book,min_support=0.2,use_colnames=True)
frequent_itemset2
```

```python
# with 60% Confidence
rules2 = association_rules(frequent_itemset2,metric='lift',min_threshold=0.6)
rules2
```

```python
# Visualizatin of Obtained rules
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
